# Univariate Analysis

```python
plt.figure(figsize=(24,30))
plt.bar(x= data ['job_title'], height= data ['salary'], color='midnightblue')
    
plt.xticks(rotation=90, fontsize=18)
plt.yticks(fontsize=18)
plt.title('salary by job title', fontsize=16, fontweight='bold')
plt.xlabel('job_title', fontsize=28)
plt.ylabel('salary', fontsize=14)
plt.show()
```

```python
# Salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['salary_in_usd'], bins=30, kde=True, color='blue')
plt.title('Salary Distribution (in USD)')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

A histogram helps identify the range of typical salaries and whether salaries are skewed (e.g., more lower-paid or higher-paid positions).
The x-axis represents salary amounts, while the y-axis represents the frequency of employees earning that salary.

```python
# Experience level distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_level', data=data, palette='Set2')
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

This bar chart displays the distribution of employees across different experience levels (e.g., Entry, Mid, Senior).
It can highlight whether the dataset is dominated by more junior or senior employees.
Companies or industries with a high number of senior-level employees tend to pay higher salaries on average.

```python
# Job title distribution (top 10)
plt.figure(figsize=(12, 8))
data['job_title'].value_counts().nlargest(10).plot(kind='barh', color='green')
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()
```

This graph helps to identify the most prevalent roles in the data science field, such as Data Scientists, ML Engineers, etc.

# Bivariate Analysis

```python
# Salary by experience level
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=data, palette='Set3')
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

This graph shows how salary varies with experience.
It can highlight differences in the salary range (minimum, maximum, and median) for each experience level.
Outliers, which are marked as dots outside the whiskers, represent unusually high or low salaries.


```python
# Salary by employment type
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=data, palette='Set1')
plt.title('Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

This boxplot shows whether certain employment types (e.g., contract vs. full-time) result in significantly different salary levels.
It can highlight which employment type typically offers higher compensation and whether there’s more salary variability within specific types.


```python
# Salary by company size
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=data, palette='Set2')
plt.title('Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.show()
```

This graph helps explore whether larger companies (typically with more resources) offer higher salaries than small or medium-sized companies.

```python
# Salary by employee residence (top 10)
plt.figure(figsize=(12, 8))
top_10_residence = data['employee_residence'].value_counts().nlargest(10).index
sns.boxplot(x='employee_residence', y='salary_in_usd', data=data[data['employee_residence'].isin(top_10_residence)], palette='Set3')
plt.title('Salary by Employee Residence (Top 10)')
plt.xlabel('Employee Residence')
plt.ylabel('Salary in USD')
plt.xticks(rotation=45)
plt.show()
```

This graph helps you see how salaries vary across different locations.
You can identify whether certain regions pay higher salaries, such as in the U.S., Europe, or Asia.

```python
# Salary by remote ratio
plt.figure(figsize=(10, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, palette='coolwarm')
plt.title('Salary by Remote Work Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary in USD')
plt.show()
```

This graph is useful for understanding how remote work impacts salary.
You can see if fully remote roles offer higher or lower salaries compared to hybrid or in-office positions, which can be influenced by the company’s location, work policies, or employee preferences.

# conclusion

Experience level and salary: Salaries tend to increase significantly with experience. Senior-level (SE) positions generally offer the highest salaries, while entry and mid-level roles earn comparatively less.

Employment type's effect on salary: Full-time employees tend to have more stable and higher salaries, while contract workers often experience greater salary variations, especially in highly specialized roles.

Company size matters: Large companies (L) typically pay more than small (S) and medium (M) companies, which could be attributed to their larger budgets and global reach.

Impact of remote work: While remote work opportunities are growing, fully remote roles (100% remote) offer competitive salaries, particularly in regions where tech talent is scarce. However, the remote ratio does not drastically influence salaries across all locations.
