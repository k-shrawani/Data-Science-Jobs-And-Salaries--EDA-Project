ok so there seems to be many outliers lets see the data without the outliers

```python
plt.figure(figsize=[16,10])
c = 1
for i in range(2020,2024):
    df2 = df[df['work_year']==i]
    Q3 = df2['salary'].quantile(0.75)
    Q1 = df2['salary'].quantile(0.25)
    IQR = Q3 - Q1

    upper = Q3 + IQR*1.5
    lower = Q1 - IQR*1.5
    non_out_df = df2[(df2['salary']>lower)&(df2['salary']<upper)]
    plt.subplot(2,2,c)
    sns.histplot(data=non_out_df,x='salary',kde=True,palette=sns.color_palette("pastel"))
    plt.title(f'{i} Salaries Distribution')
    plt.xlabel('Salaries')
    plt.ylabel('Frequency')
    c +=1
plt.show()
```

as you can see this is a better representation of the data without any outliers from this we can see

In 2020 we can see most employees salaries are 5K

In 2021 we can see most employees salaries are 7K

In 2022 we can see most employees salaries are 13K

In 2023 we can see most employees salaries are 12K
